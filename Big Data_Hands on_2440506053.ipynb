```python
from pyspark.sql import SparkSession

#Memulai Spark Session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

#Membuat DataFrame Sederhana
data = [("Ali", 34, '3D Game Artist', 'Main Game', 'L'), ("Budi", 23, 'Illustrator', 'Main Game', 'L'), ("Citra", 29, 'UI/UX Design', 'Baca Manga', 'P'), ("Dina", 45, '3D Game Artist', 'Nonton Anime', 'P')]
columns =['Nama', 'Usia', 'Pekerjaan', 'Hobi', 'Gender']
df = spark.createDataFrame(data, columns)

#Menampilkan DataFrame
df.show()

#Filtering data
print ("Filtering Data")
df_filtered = df.filter(df['Pekerjaan'] == '3D Game Artist')
df_filtered.show()

df_filtered = df.filter(df['Usia'] > 25)
df_filtered.show()

df_filtered = df.filter(df['Hobi'] == 'Main Game')
df_filtered.show()

#Menghitung rata-rata usia
print ("Rata-rata Usia")
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

#Mengurutkan data berdasarkan usia
df_sorted = df.orderBy('Usia', ascending=True)
df_sorted.show()
```

```python
from pyspark.sql import SparkSession

#Memulai Spark Session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

#Membuat DataFrame Sederhana
data = [("Ali", 34, '3D Game Artist', 'Main Game', 'L'), ("Budi", 23, 'Illustrator', 'Main Game', 'L'), ("Citra", 29, 'UI/UX Design', 'Baca Manga', 'P'), ("Dina", 45, '3D Game Artist', 'Nonton Anime', 'P')]
columns =['Nama', 'Usia', 'Pekerjaan', 'Hobi', 'Gender']
df = spark.createDataFrame(data, columns)

import pandas as pd

#Membuat DataFrame Pandas
data_pandas = {'Nama': ['Ali', 'Budi', 'Citra', 'Dina'], 'Usia': [34, 23, 29, 45]}
df_pandas = pd.DataFrame(data_pandas)


#Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

#Mengonversi DataFrame dari pandas ke pySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

#Gabungan DataFrame pandas
# Gunakan pd.concat() untuk menggabungkan DataFrame secara vertikal
df_gabungan_pandas = pd.concat([df_pandas, df_pandas_from_spark], ignore_index=True)

print('Hasil Gabungan DataFrame Pandas')
print(df_gabungan_pandas)


#Gabungan DataFrame pyspark
print('Hasil Gabungan DataFrame pyspark')
df_gabungan_pyspark = df.unionByName(df_spark_from_pandas, allowMissingColumns=True)
df_gabungan_pyspark.show()
#Menampilkan DataFrame hasel konversi


print('Hasil Konversi DataFrame dari Pandas ke PySpark')
df_spark_from_pandas.show()
print('Hasil Konversi DataFrame dari PySpark ke Pandas')
print(df_pandas_from_spark)

#Menghitung rata-rata usia

print('Rata rata usia menggunakan pyspark')
from pyspark.sql.functions import avg
df_gabungan_pyspark.groupBy().agg(avg("Usia")).show()

print('Rata rata usia menggunakan pandas')

rata_rata_usia_pandas = df_gabungan_pandas['Usia'].mean()
print(rata_rata_usia_pandas)


```

```python
#Mengonversi DataFrame dari PySpark ke Pandas
df_pandas_from_spark = df.toPandas()

#Mengonversi DataFrame dari pandas ke pySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Menampilkan DataFrame hasil konversi
print('hasil konversi spark dan pandas')
df_pandas_from_spark, df_spark_from_pandas.show()

#Gabungan DataFrame pandas
# Gunakan pd.concat() untuk menggabungkan DataFrame secara vertikal
df_gabungan_pandas = pd.concat([df_pandas, df_pandas_from_spark], ignore_index=True)
print('Hasil Gabungan DataFrame Pandas')
print(df_gabungan_pandas)

#Max usia
print('Max usia menggunakan pandas')
max_usia_pandas = df_gabungan_pandas['Usia'].max()
print(max_usia_pandas)

#Gabungan DataFrame pyspark
print('Hasil Gabungan DataFrame pyspark')
df_gabungan_pyspark = df.unionByName(df_spark_from_pandas, allowMissingColumns=True)
df_gabungan_pyspark.show()

#Max Usia
print('Max usia menggunakan pyspark')
max_usia_pyspark = df_gabungan_pyspark.agg({"Usia": "max"}).collect()[0][0]
print(max_usia_pyspark)

```
